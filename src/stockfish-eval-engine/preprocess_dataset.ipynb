```python
import chess
import torch
import pandas as pd
import math

from encoder import board_to_tensor
from dataset import evaluation_to_target

df = pd.read_csv(f"D:\Coding\CNN Chess Engine\data\chessData.csv")

```

```python
BATCH_SIZE = 250000
TOTAL_SIZE = len(df)

NUM_BATCH = math.ceil(TOTAL_SIZE/BATCH_SIZE)
START_BATCH = 20

for batch in range(START_BATCH,21):
    start_idx = (batch)*BATCH_SIZE
    end_idx = (batch+1)*BATCH_SIZE

    X=[]
    Y=[]

    for idx in range(start_idx,end_idx):

        if(idx >= TOTAL_SIZE):
            break

        row = df.iloc[idx]

        fen = row.FEN
        evaluation = row.Evaluation

        target = evaluation_to_target(evaluation)

        board = chess.Board(fen)

        tensor = board_to_tensor(board)

        X.append(tensor)
        Y.append(target)

    
    X_tensor = torch.stack(X).to(torch.uint8)
    Y_tensor = torch.tensor(Y,dtype=torch.float32)

    torch.save(X_tensor,fr"D:\Coding\CNN Chess Engine\data\pre-tensors\X_{batch+1}.pt")
    torch.save(Y_tensor,fr"D:\Coding\CNN Chess Engine\data\pre-tensors\Y_{batch+1}.pt")
    print(f"Saved : Batch {batch+1}")
```
