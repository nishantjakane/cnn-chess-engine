```python
device = torch.device("cuda")

model = ChessCNN().to(device)
checkpoint = torch.load(r"d:\Coding\CNN Chess Engine\data\chess_models\best_model.pth")
model.load_state_dict(checkpoint["model_state_dict"])

model.eval()
```

```python
X = torch.load(
    r"d:\Coding\CNN Chess Engine\data\pre-tensors\X_10.pt"
)

Y = torch.load(
    r"d:\Coding\CNN Chess Engine\data\pre-tensors\Y_10.pt"
)

dataset = TensorDataset(X,Y)
train_size = 249000
test_size = len(dataset) -train_size
train , test_dataset =  random_split(dataset,[train_size,test_size])
```

```python
fen = "2k1rb1r/pp3ppp/3p1n2/2p5/3q2b1/8/6K1/8 b - - 7 34"
board = chess.Board(fen)
target = 311
board = board_to_tensor(board)
board = board.unsqueeze(0).float().to(device)
prediction = model(board)

print(prediction)
```

```python
sum_diff =0

for board, target in test_dataset:

    board = board.unsqueeze(0).float().to(device)

    with torch.no_grad():
        prediction = model(board)
    diff = abs(prediction.item()-target.item())
    sum_diff +=diff
    print("Prediction:", prediction.item())
    print("Target:", target.item())
    print("Difference: ")

avg_diff = sum_diff/len(test_dataset)

print("Avg Difference: ",avg_diff)
```
